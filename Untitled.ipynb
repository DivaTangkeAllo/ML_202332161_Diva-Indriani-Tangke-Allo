```python
# Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Membaca DATASET

```python
df= pd.read_csv('botak.csv')
```

```python
df
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='jenis_kelamin', y='gaji', data=df, palette='Set2')
plt.title('Boxplot Gaji Berdasarkan Jenis Kelamin')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='pekerjaan', data=df, palette='Set3', order=df['pekerjaan'].value_counts().index)
plt.title('Distribusi Pekerjaan')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='provinsi', y='gaji', data=df, estimator='mean', palette='mako', order=df.groupby('provinsi')['gaji'].mean().sort_values(ascending=False).index)
plt.title('Rata-Rata Gaji Berdasarkan Provinsi')
plt.xticks(rotation=45)
plt.show()
```
